```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
study_frame=500
```

```python
def get_plots(region1, region2, data_type):
    fname1 = 'Datasets/Right_{}_{}.npy'.format(region1, data_type[0])
    fname2 = 'Datasets/Right_{}_{}.npy'.format(region2, data_type[0])
    region1_np = np.load(fname1)
    region2_np = np.load(fname2)
    
    study_frame=len(region1_np)
    
    for point_num in range(21):
        region1_data = np.zeros((1,3))
        region2_data = np.zeros((1,3))
        time_list=[]

        if data_type == "velocity":
            study_frame -=1
        elif data_type == "acceleration":
            study_frame -=2

        for i in range(study_frame):
            region1_data=np.vstack((region1_data, [[region1_np[i][point_num][0],region1_np[i][point_num][1],region1_np[i][point_num][2]]]))
            region2_data=np.vstack((region2_data, [[region2_np[i][point_num][0],region2_np[i][point_num][1],region2_np[i][point_num][2]]]))
            time_list.append(i/30)

        region1_data = np.delete(region1_data, 0, axis=0)
        region2_data = np.delete(region2_data, 0, axis=0)

        plt.figure(figsize=(15,4))
        plt.subplot(121)
        plt.grid()        
        plt.plot(time_list, region1_data[:,0], "-", label='x')
        plt.plot(time_list, region1_data[:,1], "-", label='y')
        plt.plot(time_list, region1_data[:,2], "-", label='z')
        plt.xlabel('time(s)')
        plt.ylabel(data_type)
        plt.ylim([-0.05, 0.05])
        plt.legend(loc="upper right")
        plt.title(region1+" at point:"+str(point_num))

        plt.subplot(122)
        plt.grid()        
        plt.plot(time_list, region2_data[:,0], "-", label='x')
        plt.plot(time_list, region2_data[:,1], "-", label='y')
        plt.plot(time_list, region2_data[:,2], "-", label='z')
        plt.xlabel('time(s)')
        plt.ylabel(data_type)
        plt.ylim([-0.05, 0.05])
        plt.legend(loc="upper right")
        plt.title(region2+" at point:"+str(point_num))

        plt.show()
        print()        
```

```python
get_plots("ULF", "ULB", "velocity")
```

```python
get_plots("ULF", "ULB", "acceration")
```
